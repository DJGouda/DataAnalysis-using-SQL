```python
#extracting zip file from zip file 
#(2) In this I have done for Data Collection and Storing of Data, which is the 

import zipfile
zip_reff = zipfile.ZipFile('orders.csv.zip')
zip_reff.extractall() #for changing file it to a directory
zip_reff.close() 
```

```python
#Now we will read data from orders.csv files and will handle all null values
#(3) In this we will be doing Data Cleaning, removing null values or cheking for any errors and removing it

import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#Now renaming all the column names to lowercase and replacing space with underscore

df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
#Now we will make new columns discount, sale price and profit

df['discount']=df['list_price']*df['discount_percent']*0.01
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
#Now we will convert date from object data type to datetime

df['order_date']=pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df
```

```python
#Now we will drop the cost price, list price and discount percentage column

df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df
```

```python
#load the data into mysql server 
import sqlalchemy as sal
engine = sal.create_engine('mysql://duren\MySQL/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn=engine.connect()
```

```python
#load the data into mysql server using append option
df.to_sql('df_orders', con=conn , index=False, if_exists = 'append')
```
